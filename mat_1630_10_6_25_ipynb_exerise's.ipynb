# list comprehension **approach**

```python
f_list = [4*n**3 - 1 for n in range(7)]
print(f_list)
```

# Numpy approach

```python
import numpy as np

x = np.arange(7)
f_np = 4*x**3 - 1
print(f_np)
```

# list comprehension part 2

```python
S = [(m, n) for m in range(4) for n in range(4)]
print(S)
print(len(S))
```

# numpy arrays practice

```python
import numpy as np

L = [2, 4, 6, 8, 10]
a = np.array(L)
print("a =", a)
print("a**2 =", a**2)
print("cumsum =", a.cumsum())
print("sin(a) =", np.sin(a))
x = np.linspace(0, 1, 9)
print("linspace =", x)
M = np.array([
    [1, 2, 3],
    [0, 0, 0],
    [7, 5, 4]
])
print("M =\n", M)
print("M[0,1] =", M[0,1])      # first row, second column
print("M[:,2] =", M[:,2])      # third column
print("M[1:,:] =\n", M[1:,:])  # first row as 2D slice
vstacked = np.stack((a, a**2))
hstacked = np.hstack((a, a**2))

print("Vertical stack:\n", vstacked)
print("Horizontal stack:\n", hstacked)
```
